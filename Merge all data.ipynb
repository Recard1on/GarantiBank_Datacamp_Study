```python
## I will read and combine all the additional data here
```

```python
dil=pd.read_csv("dil.csv")
edu = pd.read_csv("edu.csv")
skil = pd.read_csv("skil.csv")
work = pd.read_csv("is.csv")
test = pd.read_csv("test_users.csv")
sub = pd.read_csv("submission.csv")
```

```python
train = train.set_index("user_id")
test = test.set_index("user_id")
sub = sub.set_index("user_id")
```

```python
train[edu.columns]   = edu[edu.columns]
train[dil.columns]  = dil[dil.columns]
train[skil.columns] = skil[skil.columns]
train[work.columns]   = work[work.columns]

test[edu.columns]   = edu[edu.columns]
test[dil.columns]  = dil[dil.columns]
test[skil.columns] = skil[skil.columns]
test[work.columns]   = work[work.columns]
```

```python
train['(1th)company_id'] = train['(1th)company_id'].astype('object')
train['(2th)company_id'] = train['(2th)company_id'].astype('object')
train['(3th)company_id'] = train['(3th)company_id'].astype('object')
train['(4th)company_id'] = train['(4th)company_id'].astype('object')
test['(1th)company_id'] = test['(1th)company_id'].astype('object')
test['(2th)company_id'] = test['(2th)company_id'].astype('object')
test['(3th)company_id'] = test['(3th)company_id'].astype('object')
test['(4th)company_id'] = test['(4th)company_id'].astype('object')
```

```python
train
```

```python
### I filled the empty data in all data with the most repeating value of that column
```

```python
num_cols = train.select_dtypes(include='number').columns.tolist()
train[num_cols] = train[num_cols].fillna(train[num_cols].mode().iloc[0])
```

```python
cat_cols = train.select_dtypes(include=['object']).columns
train[cat_cols] = train[cat_cols].fillna(train[cat_cols].mode().iloc[0])
```

```python
numeric_cols = test.select_dtypes(include='number').columns.tolist()
test[numeric_cols] = test[numeric_cols].fillna(test[numeric_cols].mode().iloc[0])
cat_cols = test.select_dtypes(include=['object']).columns
test[cat_cols] = test[cat_cols].fillna(test[cat_cols].mode().iloc[0])
```

```python
pd.options.display.max_columns = 200
train.info()
```

```python
test.info()
```

```python
train.drop("user_id",axis=1,inplace=True)
test.drop("user_id",axis=1,inplace=True)
train["moved_after_2019"]=train["moved_after_2019"].astype("int64")
train
```

```python
### Modelling
```

```python
X = train.drop(["moved_after_2019"],axis=1)
y = train["moved_after_2019"]
```

```python
cat_features = ["industry", "location", "graduate", "master", "other_degree", "phd",
                "(1th)company_id", "(2th)company_id", "(3th)company_id","(4th)company_id","(1th)company_location",
                "(2th)company_location", "(3th)company_location","(4th)company_location"]
X[cat_features] = X[cat_features].astype(str)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 15)
```

```python
from sklearn.model_selection import KFold
from catboost import CatBoostClassifier
```

```python
## I used catboost algorithm and validated with k fold and trained my model 
## and experimented with test data and filled the sub misson file
```

```python
catboost_model = CatBoostClassifier(cat_features=cat_features)
kf = KFold(n_splits=6)
for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    catboost_model.fit(X_train, y_train)
    print(catboost_model.score(X_test, y_test))
```

```python
test[cat_features] = test[cat_features].astype(str) # bu satırı ekleyin
catboost_model = CatBoostClassifier(cat_features=cat_features)
catboost_model.fit(X_train, y_train)
sub.loc[test.index, "moved_after_2019"] = catboost_model.predict(test)
sub["moved_after_2019"] = sub["moved_after_2019"].astype(int)
sub["moved_after_2019"].value_counts()
```

```python
sub.to_csv('cat_sub.csv')
```
