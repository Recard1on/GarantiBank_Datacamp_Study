```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
idz="user_id"
```

```python
work = pd.read_csv("work_experiences.csv")
```

```python
work.info()
```

```python
## Since the data class after this date is 1, I got the previous data.
```

```python
work = work[work["start_year_month"]<201901]
```

```python
df_exp = work.sort_values(by=['user_id', 'start_year_month'])
df_exp
```

```python
### I scanned the company ID and filled in the location of that company ID and the blank location information.
```

```python
missing_locations = df_exp[df_exp['location'].isnull()]['company_id']

for company_id in missing_locations:
    first_valid_location = df_exp.loc[df_exp['company_id'] == company_id, 'location'].dropna().iloc[0] if len(df_exp.loc[df_exp['company_id'] == company_id, 'location'].dropna()) > 0 else None
    if first_valid_location is not None:
        df_exp.loc[(df_exp['company_id'] == company_id) & (df_exp['location'].isnull()), 'location'] = first_valid_location
```

```python
df_exp
```

```python
df_new=pd.DataFrame()
```

```python
### Since there is repetitive data, I created it in the pivot table logic as follows
```

```python
df_new['(1th)company_id'] = df_exp.groupby(idz)['company_id'].nth(-1).astype(str)
df_new['(2th)company_id'] = df_exp.groupby(idz)['company_id'].nth(-2).astype(str)
df_new['(3th)company_id'] = df_exp.groupby(idz)['company_id'].nth(-3).astype(str)
df_new['(4th)company_id'] = df_exp.groupby(idz)['company_id'].nth(-4).astype(str)

df_new['(1th)company_location'] = df_exp.groupby(idz)['location'].nth(-1).astype(str)
df_new["(2th)company_location"] = df_exp.groupby(idz)['location'].nth(-2).astype(str)
df_new['(3th)company_location'] = df_exp.groupby(idz)['location'].nth(-3).astype(str)
df_new["(4th)company_location"] = df_exp.groupby(idz)['location'].nth(-4).astype(str)

df_new['min_exp_time'] = df_exp.groupby(idz)['start_year_month'].min()
df_new['max_exp_time'] = df_exp.groupby(idz)['start_year_month'].max()

df_new
```

```python
df_new = df_new.replace('nan', np.nan)
```

```python
### I filled the blank data with the last company id and company location.
```

```python
df_new['(2th)company_id'].fillna(df_new['(1th)company_id'],inplace=True)
df_new['(3th)company_id'].fillna(df_new['(2th)company_id'],inplace=True)
df_new['(4th)company_id'].fillna(df_new['(3th)company_id'],inplace=True)
df_new['(2th)company_location'].fillna(df_new['(1th)company_location'],inplace=True)
df_new['(3th)company_location'].fillna(df_new['(2th)company_location'],inplace=True)
df_new['(4th)company_location'].fillna(df_new['(3th)company_location'],inplace=True)
df_new
```

```python
df_new.isnull().sum()
```

```python
### I filled the remaining blank data as -1 as numeric values ​​and other as categorical data
```

```python
df_new[["(2th)company_id","(3th)company_id","(4th)company_id"]] = df_new[["(2th)company_id","(3th)company_id",
                                                                          "(4th)company_id"]].fillna("-1")
df_new[["(1th)company_location","(2th)company_location","(3th)company_location","(4th)company_location"]] = df_new[["(1th)company_location",
                                                                                                                    "(2th)company_location","(3th)company_location","(4th)company_location"]].fillna("other")
df_new
```

```python
df_new.to_csv("is.csv")
```

```python
dil = pd.read_csv("languages.csv")
```

```python
dil["language"] = dil["language"].str.lower()
dil["language"] = dil["language"].str.capitalize()
dil
```

```python
### I tried to reduce the same category to a single class because it is spelled differently.
```

```python
dil["language"] = dil["language"].replace('İngilizce','English')
dil["language"] = dil["language"].replace('Ingilizce','English')
dil["language"] = dil["language"].replace('Türkçe', 'Turkish')
dil["language"] = dil["language"].replace('Fransızca', 'French')
dil["language"] = dil["language"].replace('Almanca', 'German')
dil["language"] = dil["language"].replace('İspanyolca', 'Spanish')
dil["language"] = dil["language"].replace('Arapça', 'Arabic')
dil["language"] = dil["language"].replace('İtalyanca', 'Italian')
dil["language"] = dil["language"].replace("Rusça", 'Russian')
dil
```

```python
dil = dil.sort_values(by=["user_id"])
```

```python
### I changed language levels with numeric variables with tag encoding logic
```

```python
dil["proficiency"] = dil["proficiency"].replace("elementary", '1')
dil["proficiency"] = dil["proficiency"].replace("limited_working", '2')
dil["proficiency"] = dil["proficiency"].replace("professional_working", "3")
dil["proficiency"] = dil["proficiency"].replace("full_professional", '4')
dil["proficiency"] = dil["proficiency"].replace("native_or_bilingual", '5')
dil.fillna(1,inplace=True)
```

```python
### I determined the 8 most repetitive languages and changed the ones that were not among these languages 
### to other, and then combined them with the language levels.
```

```python
top_langs = dil["language"].value_counts().nlargest(8).index.tolist()
dil['language'] = dil['language'].apply(lambda x: x if x in top_langs else 'other')
dil
```

```python
dil = dil.drop_duplicates(['user_id', 'language'])
dil = pd.pivot(dil, index='user_id', columns='language', values='proficiency')
dil = dil.fillna(0).astype(int)
dil
```

```python
dil.to_csv("dil.csv")
```

```python
skil = pd.read_csv("skills.csv")
```

```python
### To identify the 60 most repetitive skills. I updated the remaining abilities as other.
### Then I replaced it with one hot encoding logic
```

```python
top_skill = skil["skill"].value_counts().nlargest(60).index.tolist()
skil.loc[~skil["skill"].isin(top_skill), "skill"] = "other_skil"
skil
```

```python
skil["deneyim"]=True
skil = skil.drop_duplicates(['user_id', 'skill'])
skil = pd.pivot(skil, index='user_id', columns='skill', values='deneyim')
skil = skil.fillna(0).astype(int)
skil
```

```python
skil.to_csv("skil.csv")
```

```python
edu = pd.read_csv("education.csv")
```

```python
edu
```

```python
edu.drop(["start_year_month","end_year_month"],axis=1,inplace=True)
```

```python
edu.isnull().sum()
```

```python
### There are data from the same category but with different spellings, I converted them into a single category.
```

```python
edu_conv = {
    "(?i).*(doktora|doctor|phd|ph.d).*": "phd",
    "(?i).*(yüksek|master|M.Sc|MS|MBA|M.S).*": "master",
    "(?i).*(lisans|licence|bachelor|BS|B.S|B.Sc|engineer|mühendis|mezun|graduate|licentiate).*": "graduate",
}
edu['degree'] = edu['degree'].replace(edu_conv, regex=True)
edu
```

```python
edu = edu.sort_values(by=["user_id","degree"])
```

```python
edu.drop("fields_of_study",axis=1,inplace=True)
```

```python
edu
```

```python
### I determined the 3 most repetitive degrees again and updated the remaining ones as other.
```

```python
edu.fillna("graduate",inplace=True)
top_edu = edu["degree"].value_counts().nlargest(3).index.tolist()
edu.loc[~edu["degree"].isin(top_edu), "degree"] = "other_degree"
edu = edu.drop_duplicates(['user_id', 'degree'])
edu = pd.pivot(edu, index='user_id', columns='degree', values='school_name')
edu
```

```python
edu.fillna("unkown",inplace=True)
```

```python
edu.info()
```

```python
edu.to_csv("edu.csv")
```
